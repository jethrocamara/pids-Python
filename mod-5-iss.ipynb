```python
from urllib.request import Request 
from urllib.request import urlopen
import json

req = Request("http://api.open-notify.org/iss-now.json")
response = urlopen(req)

obj = json.loads(response.read())

print(obj['timestamp'])
print(obj['iss_position']['latitude'])
print(obj['iss_position']['longitude'])
```

```python
import csv
import matplotlib.pyplot as plt

x = []
y = []

with open('ISS Ping.csv', 'r') as csv_file:
    iss = csv.reader(csv_file, delimiter=',')
    
    next(iss)

    for row in iss:
        x.append(row[1])
        y.append(row[2])

plt.scatter(x,y)
```
